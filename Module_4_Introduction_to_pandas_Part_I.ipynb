# **Series**

A Series in Pandas is like a simple list or array, but with extra features. It has two main parts:

- Data array – This holds the actual values (like numbers or text - data of any numpy type).

- Index array – This is a set of labels that helps you identify each value in the data array.

**Declaring a Series**

To create a Series, use the pd.Series() constructor and pass an array of values to it. For example:

```python
import pandas as pd

s = pd.Series([12, -4, 7, 9])
print(s)
```

In the output, the left side shows the index (which are labels), and the right side shows the values. By default, if you don’t specify an index, Pandas uses numerical labels starting from 0.

**Using Custom Labels**

To make the Series more meaningful, you can specify custom labels (index). Here's how:

```python
s = pd.Series([12, -4, 7, 9], index=['a', 'b', 'c', 'd'])
print(s)
```

Accessing **index and Values**

To see the index or the values separately, use the index and values attributes:

```python
print(s.values)  # Output: [12 -4  7  9]
print(s.index)   # Output: Index(['a', 'b', 'c', 'd'], dtype='object')
```

**Summary**
- A Series is created by passing a list or array of values.
- If you don't specify an index, Pandas automatically uses numbers starting from 0.
- You can provide custom labels for clarity.
- The .values attribute gives you the data array, and the .index attribute gives you the labels.

**Selecting Elements from a Series**

You can select elements from a Series in the same way you would from a NumPy array, using their position or label.


**Selecting by Position:**

```python
import pandas as pd

s = pd.Series([12, -4, 7, 9], index=['a', 'b', 'c', 'd'])
print(s[2])  # Output: 7
```

This selects the element at position 2 (counting from 0).

**Selecting by Label:**

```python
print(s['b'])  # Output: -4
```

This selects the element with the label 'b'.


**Selecting Multiple Elements**

Using Position Ranges:

```python
print(s[0:2])
```

This selects elements from position 0 up to, but not including, position 2.

**Using a List of Labels:**

```python
print(s[['b', 'c']])
```

This selects the elements with labels 'b' and 'c'.

**Summary**
- Use s[position] to select by position.
- Use s['label'] to select by label.
- For multiple selections, use s[start:end] or s[['label1', 'label2']].

**Assigning Values to Elements in a Series**

Once you know how to select elements, you can also change their values. You can assign new values to elements using either their position or their label.

```python
import pandas as pd

s = pd.Series([12, -4, 7, 9], index=['a', 'b', 'c', 'd'])
s[1] = 0  # Assigning 0 to the element at position 1
print(s)
```

This changes the value at position 1 (label 'b') to 0.

**Assigning by Label:**

```python
s['b'] = 1  # Assigning 1 to the element with label 'b'
print(s)
```

This changes the value of the element with the label 'b' to 1.

**Summary**
- You can modify the value of an element in a Series by using its position (s[1] = value) or its label (s['label'] = value).
- The changes are directly reflected in the Series.

**Defining Series from NumPy Arrays and Other Series**

You can create new Series from NumPy arrays or existing Series. Here’s how:

Creating a Series from a NumPy Array:

```python
import pandas as pd
import numpy as np

arr = np.array([1, 2, 3, 4])
s3 = pd.Series(arr)
print(s3)
```

This creates a Series s3 using the values from the NumPy array arr.

**Creating a Series from Another Series:**

```python
s = pd.Series([12, 4, 7, 9], index=['a', 'b', 'c', 'd'])
s4 = pd.Series(s)
print(s4)
```

This creates a new Series s4 with the same values and labels as the original Series s.

**Important Note on References**

When you create a Series from a NumPy array or another Series, the values are not copied but passed by reference. This means changes to the original array or Series will reflect in the new Series.

```python
arr[2] = -2  # Change the third element of arr
print(s3)
```

As shown, changing arr[2] to -2 also changed the third element in s3 because s3 is referencing arr directly.

**Summary**
- You can create Series using NumPy arrays or existing Series.
- Be aware that changes to the original array or Series will be reflected in the new Series because they share the same reference.







**Filtering Values in a Series**

Pandas is built on top of the NumPy library, so many operations that work with NumPy arrays also work with Series. One such operation is filtering values based on conditions.

**Example**: Filtering with a Condition


Suppose you have a Series and you want to find out which elements have a value greater than 8. You can use a condition to filter the Series:

```python
import pandas as pd

s = pd.Series([12, 4, 7, 9], index=['a', 'b', 'c', 'd'])
print(s>8)
print(s[s > 8])
```

**Explanation**

- s > 8 creates a Boolean mask that is True for elements where the condition is met (i.e., elements greater than 8) and False otherwise.
- s[s > 8] uses this mask to filter and return only the elements where the condition is True.

**Summary**

- You can filter a Series using conditions, just like with NumPy arrays.
- This operation helps you extract specific data based on conditions (e.g., values greater than, less than, or equal to a certain number).

**Operations and Mathematical Functions in a Series**

Pandas Series support many operations and mathematical functions similar to those used with NumPy arrays. You can use both basic arithmetic operators and NumPy functions on Series objects.

Using Arithmetic Operators: You can perform arithmetic operations directly on Series using standard operators like +, -, *, and /.

```python
import pandas as pd

s = pd.Series([12, -4, 7, 9], index=['a', 'b', 'c', 'd'])
print(s / 2)
```

Here, every element in the Series is divided by 2.

**Using NumPy Mathematical Functions:**

For more complex mathematical operations, you need to use NumPy functions, which should be called with np. and the Series passed as an argument.

Example: Applying the Log Function

```python
import numpy as np

print(np.log(s))
```

- The np.log() function calculates the natural logarithm for each value in the Series.
- If a value is zero or negative, the result will be NaN (Not a Number).


**Summary**
- Arithmetic operations like +, -, *, / can be used directly on Series.
- NumPy functions (e.g., np.log(), np.sqrt()) must be called with the np. prefix and applied to the Series.

**Evaluating Values in a Series**

A Series can contain duplicate values, and you might need to check what values are present, count duplicates, or verify if specific values are in the Series.

**Example: Creating a Series with Duplicates**

```python
import pandas as pd

serd = pd.Series([1, 0, 2, 1, 2, 3], index=['white', 'white', 'blue', 'green', 'green', 'yellow'])
print(serd)
```

**Key Functions for Evaluating Values**

1. unique() Function:

    - Returns an array of unique values from the Series (not necessarily in order).

```python
print(serd.unique())
```

2. value_counts() Function:

    - Counts the number of occurrences of each unique value in the Series.


```python
print(serd.value_counts())
```

This output shows that 2 and 1 each appear twice, while 3 and 0 appear once.

3. isin() Function:

    - Checks whether each element in the Series is part of a given list of values, returning a Series of Boolean values.

```python
print(serd.isin([0, 3]))
```

Filtering Based on Membership: You can use isin() to filter the Series to show only values that match certain criteria.

```python
print(serd.isin([1,0]))
print(serd[serd.isin([1,0])])
```

Summary
- unique() gives you all unique values in a Series.
- value_counts() counts the frequency of each unique value.
- isin() checks if elements in the Series are part of a specified list and returns a Boolean mask. This can be used for filtering the Series.

**NaN Values in a Series**

In pandas, NaN (Not a Number) is used to represent missing or undefined data. This can occur when data is missing from a source or as a result of certain operations that produce undefined results, such as taking the logarithm of a negative number.

```python
import pandas as pd
import numpy as np

s2 = pd.Series([5, -3, np.NaN, 14])
print(s2)
```

**Identifying NaN Values**

To work with NaN values, pandas provides the functions isnull() and notnull():

- isnull(): Returns True for NaN values and False otherwise.
- notnull(): Returns True for values that are not NaN and False for NaN.

```python
print(s2.isnull())
```

```python
print(s2.notnull())
```

**Filtering Based on NaN Values**

You can use these functions to filter out NaN values or keep only the non-NaN values.

Example: Filtering Non-NaN Values:

```python
print(s2[s2.notnull()])
```

**Example: Filtering Only NaN Values:**

```python
print(s2[s2.isnull()])
```

**Summary**
- NaN represents missing or undefined data in a Series.
- Use isnull() to find NaN values and notnull() to find non-NaN values.
- These functions can be used in filtering to selectively keep or remove NaN values from a Series.

**Series as Dictionaries**

A Series in pandas can be thought of as similar to a dictionary. In fact, you can create a Series directly from a dictionary. The keys of the dictionary become the index labels in the Series, and the corresponding values become the data in the Series.

**Example 1: Creating a Series from a Dictionary**

```python
mydict = {'red': 2000, 'blue': 1000, 'yellow': 500, 'orange': 1000}
myseries = pd.Series(mydict)
print(myseries)
```

- The keys from mydict ('red', 'blue', 'yellow', 'orange') become the index labels.
- The values in mydict (2000, 1000, 500, 1000) become the data in the Series.

**Example 2: Specifying Index Manually**

You can also create a Series with a specific order of index labels by passing an index argument. If the index contains labels that don't exist in the dictionary, NaN (missing value) is added for those labels.

```python
colors = ['red', 'yellow', 'orange', 'blue', 'green']
myseries = pd.Series(mydict, index=colors)
print(myseries)
```

- Here, the index ['red', 'yellow', 'orange', 'blue', 'green'] is specified.
- The label 'green' is not in mydict, so its value in the Series is NaN.

**Operations between Series**

You can perform arithmetic operations between two Series in pandas, similar to operations between a Series and a scalar. When you perform these operations between Series, the labels are important, and the data is aligned based on matching labels.

```python
mydict = {'red': 2000, 'blue': 1000, 'yellow': 500, 'orange': 1000}
myseries = pd.Series(mydict)

mydict2 = {'red': 400, 'yellow': 1000, 'black': 700}
myseries2 = pd.Series(mydict2)

result = myseries + myseries2
print(result)
```


- The operation myseries + myseries2 adds the values in the Series where labels match ('red' and 'yellow').
- For labels that do not match, such as 'blue', 'orange', 'green', and 'black', the result is NaN (missing value).
- Only elements with the same label are added together, so:
    - 'red': 2000 (from myseries) + 400 (from myseries2) = 2400
    - 'yellow': 500 (from myseries) + 1000 (from myseries2) = 1500

## **The DataFrame**

In a Series, there’s a single index (a list of labels) that connects each value to its label.

In a DataFrame, there are two indexes:

Row index: Similar to the Series index, this links each row to its label.
Column index: Labels each column, where each column acts like its own Series.


You can think of a DataFrame as a dictionary of Series:

- The keys are the column names (labels).
- The values are the Series (columns) in the DataFrame.
- Each Series uses the same row index to align its elements with other Series in the DataFrame.


Screen Shot 2024-12-05 at 19.03.23.png

**Defining a DataFrame**

To create a DataFrame, the most common method is to use a dictionary.

- Each key in the dictionary becomes a column name in the DataFrame.
- The values for each key are arrays (lists) that become the data in the corresponding column.

```python
import pandas as pd

# Define the data dictionary
data = {
    'color': ['blue', 'green', 'yellow', 'red', 'white'],
    'object': ['ball', 'pen', 'pencil', 'paper', 'mug'],
    'price': [1.2, 1.0, 0.6, 0.9, 1.7]
}

# Create a DataFrame
frame = pd.DataFrame(data)

# Display the DataFrame
print(frame)
```

**Selecting Specific Columns in a DataFrame**

- When creating a DataFrame from a dictionary with more data than needed, you can choose which columns to include.
- The columns option in the pd.DataFrame() constructor allows you to specify which columns to create.
- The columns will appear in the order you provide, regardless of their order in the original dictionary.

```python
data = {
    'color': ['blue', 'green', 'yellow', 'red', 'white'],
    'object': ['ball', 'pen', 'pencil', 'paper', 'mug'],
    'price': [1.2, 1.0, 0.6, 0.9, 1.7]
}

frame2 = pd.DataFrame(data, columns=['object', 'price'])
print(frame2)
```

- The data dictionary contains columns 'color', 'object', and 'price'.
- We want to create a DataFrame but only with the 'object' and 'price' columns.
- Using columns=['object', 'price'], we ensure only these columns appear in the DataFrame, in that order.

- If labels are not specified, pandas assigns default numeric labels (0, 1, 2, ...).
- To specify custom labels for the index, use the index option in the pd.DataFrame() constructor.
- DataFrames can also be created using a matrix of values along with arrays for custom row and column labels.

**Assigning Custom Index Labels**

```python
data = {
    'color': ['blue', 'green', 'yellow', 'red', 'white'],
    'object': ['ball', 'pen', 'pencil', 'paper', 'mug'],
    'price': [1.2, 1.0, 0.6, 0.9, 1.7]
}

frame2 = pd.DataFrame(data, index=['one', 'two', 'three', 'four', 'five'])
print(frame2)
```

- The frame2 DataFrame is created from the data dictionary with custom index labels ('one', 'two', 'three', 'four', 'five').
- If the index option is not used, pandas would have assigned numeric labels starting from 0.

**Creating DataFrame from a Matrix**

```python
import numpy as np

frame3 = pd.DataFrame(np.arange(16).reshape((4, 4)),
                      index=['red', 'blue', 'yellow', 'white'],
                      columns=['ball', 'pen', 'pencil', 'paper'])
print(frame3)
```

- np.arange(16).reshape((4, 4)) creates a 4x4 matrix filled with numbers from 0 to 15.
- The index and columns options assign custom labels to the rows and columns, respectively.

**Checking Membership with isin():**

- The isin() function checks whether elements in a DataFrame belong to a specified list of values.
- It returns a new DataFrame of Boolean values, where True indicates membership.

```python
# Checking membership in the DataFrame
result = frame.isin([1.0, 'pen'])
print(result)
```

**Filtering Rows Based on Membership:**

- You can use the Boolean DataFrame returned by isin() as a condition to filter the original DataFrame.
- When applied, only the matching elements will appear, while all others are replaced with NaN.

```python
# Filtering the DataFrame based on membership
filtered_frame = frame[frame.isin([1.0, 'pen'])]
print(filtered_frame)
```

**Key Notes:**

1. isin() with DataFrames:
    - Useful for verifying if multiple values exist in any part of the DataFrame.
2. Filtering Using Membership:

**Deleting a Column in a DataFrame:**


- To remove an entire column along with its data, you can use the del statement.
- This operation modifies the DataFrame in place, permanently deleting the specified column.

```python
# Assume 'frame' is the DataFrame
print("Before deleting the column:")
print(frame)

# Delete the column named 'new'
del frame['new']

print("\nAfter deleting the column:")
print(frame)
```

**Key Points:**

1. del Statement:
    - Directly deletes the specified column from the DataFrame.
2. Permanent Modification:
    - Once deleted, the column is removed permanently unless you have another reference to it or saved it elsewhere.
3. Alternative:

**Filtering Rows Based on Conditions:**

- In a DataFrame, you can apply filtering to extract specific rows or values based on a condition.
- When a condition is applied, a new DataFrame is returned, where:
    - The values meeting the condition remain as is.
    - The values not meeting the condition are replaced with NaN.

```python
# Example DataFrame
import pandas as pd
import numpy as np

data = {
    'ball': [0, 4, 8, np.nan],
    'pen': [1, 5, 9, np.nan],
    'pencil': [2, 6, 10, np.nan],
    'paper': [3, 7, 11, np.nan]
}
frame = pd.DataFrame(data, index=['red', 'blue', 'yellow', 'white'])

# Filtering: Values less than 12
filtered_frame = frame[frame < 12]

print("Filtered DataFrame:")
print(filtered_frame)
```

**Key Points:**

1. Condition-Based Filtering:
You can apply any logical condition (e.g., <, >, ==, etc.) to filter values.
2. Original Positions Retained:
The filtered DataFrame keeps the original structure, with unqualified values replaced by NaN.
3. Common Use Cases:
Extracting rows/columns based on numerical thresholds or specific conditions.
4. Further Customization:
Combine filtering with logical operators for more complex queries (e.g., (frame['ball'] < 5) & (frame['pen'] > 2)).

**Creating a DataFrame from a Nested Dictionary**

- Pandas can interpret a nested dictionary to construct a DataFrame. In this context:

- Outer Dictionary Keys: These become column names.
- Inner Dictionary Keys: These are treated as index labels.
- Inner Dictionary Values: These populate the corresponding cells in the DataFrame.
- Missing Values: If an index/column pair is not defined in the nested dictionary, Pandas fills it with NaN.

```python
import pandas as pd

# Nested dictionary
nestdict = {
    'red': {2012: 22, 2013: 33},
    'white': {2011: 13, 2012: 22, 2013: 16},
    'blue': {2011: 17, 2012: 27, 2013: 18}
}

# Creating a DataFrame
frame2 = pd.DataFrame(nestdict)

print("Resulting DataFrame:")
print(frame2)
```

Key Points:
1. Handling Missing Values:

    - The year 2011 does not exist in the red dictionary, so the value in the red column for index 2011 is NaN.
2. Automatic Alignment:

    - Pandas automatically aligns the inner dictionary keys (index labels) across all columns.
3. Customizing Columns or Indexes:

    - You can specify column or index orders explicitly when creating the DataFrame, using the columns or index arguments.


**Specifying Columns or Index**

To enforce a specific order or include additional columns/indexes:


```python
frame2 = pd.DataFrame(nestdict, index=[2011, 2012, 2013, 2014])
print(frame2)
```

Transposing a DataFrame in Pandas is straightforward using the .T attribute. Transposition flips the rows and columns, turning columns into rows and rows into columns.

```python
import pandas as pd

# Define a nested dictionary and create a DataFrame
nestdict = {
    'red': {2012: 22, 2013: 33},
    'white': {2011: 13, 2012: 22, 2013: 16},
    'blue': {2011: 17, 2012: 27, 2013: 18}
}

frame2 = pd.DataFrame(nestdict)

# Transpose the DataFrame
transposed_frame = frame2.T

# Display the transposed DataFrame
print("Original DataFrame:")
print(frame2)
print("\nTransposed DataFrame:")
print(transposed_frame)
```

**Key Points:**

- The .T attribute performs a simple, quick transposition.
- Notice how:
    - Columns in the original DataFrame become rows in the transposed DataFrame.
    - Rows in the original DataFrame become columns in the transposed DataFrame.
- Missing values (NaN) are preserved during transposition.

This operation is particularly useful when you need to switch perspectives of your data for analysis or visualization purposes.

The Index object in Pandas is a fundamental part of the library, providing structure and labels for data in Series and DataFrames. It helps organize and align data during operations, ensuring data consistency and making many operations efficient.

**Key Features of the Index Object**

**Labels for Axes:**

The Index provides labels for rows in a Series or rows and columns in a DataFrame.

```python
import pandas as pd

ser = pd.Series([5, 0, 3, 8, 4], index=['red', 'blue', 'yellow', 'white', 'green'])
print("Series:\n", ser)
print("Index:", ser.index)
```

**Immutable Nature:**

- Once created, an Index is immutable. This ensures stability, especially when Index objects are shared between different data structures.
- Attempting to modify the Index directly raises an error:


```python
ser.index[0] = 'purple'  # Raises a TypeError
```

**Metadata Handling:**

Index objects can store metadata like axis names, helping clarify data contexts.

The **idxmin**() and **idxmax**() methods in pandas are used to identify the index label corresponding to the minimum and maximum values in a Series (or DataFrame along a specified axis). These methods are particularly useful for quick insights into the data.


```python
import pandas as pd

# Create a Series
ser = pd.Series([5, 0, 3, 8, 4], index=['red', 'blue', 'yellow', 'white', 'green'])

# Get the index of the minimum and maximum values
min_index = ser.idxmin()  # Index of minimum value
max_index = ser.idxmax()  # Index of maximum value

print("Series:")
print(ser)
print("\nIndex of minimum value:", min_index)
print("Index of maximum value:", max_index)
```

```python
# Create a DataFrame
data = {
    'A': [10, 20, 15],
    'B': [30, 25, 35],
    'C': [50, 45, 40]
}
df = pd.DataFrame(data, index=['row1', 'row2', 'row3'])

# Get the index of min/max values along columns
min_idx = df.idxmin(axis=0)  # Minimum value index for each column
max_idx = df.idxmax(axis=0)  # Maximum value index for each column

print("DataFrame:")
print(df)
print("\nIndex of minimum values for each column:")
print(min_idx)
print("\nIndex of maximum values for each column:")
print(max_idx)
```

**Methods and Properties:**

.values: Returns the underlying data of the Index.

.dtype: Displays the data type of the Index.

.is_unique: Checks if the Index contains unique values.

```python
print("Values:", ser.index.values)
print("Data Type:", ser.index.dtype)
print("Is Unique:", ser.index.is_unique)
```

**Interoperability:**

Indexes can be shared or reused across different Series or DataFrames.

**Operations with Index**

Checking Membership:


```python
print('red' in ser.index)  # True
print('purple' in ser.index)  # False
```

**Converting to a List:**

```python
index_list = list(ser.index)
print(index_list)  # ['red', 'blue', 'yellow', 'white', 'green']
```

**Resetting Index:**
 To convert an Index back to a column or reset it entirely

```python
df = ser.reset_index()
print(df)
```
