## Control Flow Statements in Python
Control flow structures in Python are essential for directing the execution of a program based on certain conditions and loops. These structures allow for decision-making and repetition, enabling more complex and functional programs. Here’s an overview of the primary control flow structures in Python:
  ## 1. Decision - Making Structures
Decision - making structures require the programmer to specify one or more conditions to be evaluated or tested by the program, along with a statement or statements to be executed if the condition is determined to be true, and optionally, other statements to be executed if the condition is detetmined to be false.

- The "if" structure is the general form of a typical decision-making structure found in most of the programming languages.
-There are 4 types of "if" statements that we can use in Python programming based on the requirement:
1.1 "if" Statement
  - A simple "if" statement consists of a Boolean expression followed by one or more statements.
  - If the Boolean expression evaluates to be true, then the block of code inside the "if" statement will be executed.
  - If the condition is false, then the statements inside the "if" statement body are completely ignored

```python
  if (boolean_expression):
    statement (s) will execute if the boolean expression is true
```

1.4 if-elif-else Ladder
- An "if" statement can be followed by an optional "else if...else" statement, which is very useful to test various conditions using single "if...else if" statement. Here, a user can decide among multiple options.
- The "if" statements are executed from the top-down.
- As soon as one of the conditions controlling the "if" is true, the statement associated with that "if" is executed, and the rest of the ladder is bypassed.
- If none of the conditions are true, then the final "else" statement will be executed
- When using "if, else if, else" statements, there are few points to keep in mind.
  - An "if" can have zero or one "else" and it must come after any "else if's".
  - An "if" can have zero to many "else if's" and they must come before the "else".
  - Once an "else if" succeeds, none of the remaining else if's or "else's" will be tested.

Syntax
```python
if boolean_expression 1:
  Executes when the boolean expression 1 is true.
elif boolean_expression 2:
  Executes when the boolean expression 2 is true.
............
else:
  Executes when none of the above condition is true.

- end denotes the ending number
- step is the optional step for increment or decrement (default are +1 and −1)

##**While Loop**
- The “while” loop executes the same code again and again until a stop condition is met.
- A condition is tested in starting of “while” loop. If it is true, then loop body is executed.
- Once the loop body is executed, the condition is tested again, and so forth, until the condition is false, after
which the loop exits. This means that when the number of loops is not known in advance, we use a “while” loop.
- A situation that might occur in a “while” loop is that the loop might not run ever. When the condition is tested and the result is false, the loop body will be skipped and the first statement after the while loop will be executed. - Besides, the programmer needs to be careful that the counting variable within the loop is incremented, otherwise an infinite loop occurs.

Syntax
```python
while (condition):
  commands to be executed as long as condition is TRUE
```

###**Abnormal Loop Termination**
- Loops execute until their condition becomes false.
- Condition is checked at the beginning of each iteration.
- If the condition becomes false mid-loop, the loop doesn’t exit immediately.
- Normally, loops execute all statements within the body.
- In rare scenarios, loops need to exit prematurely.
- Solution: Use control statements like break, pass, and continue.

**Loop Control Statements**
- break: Exits the loop immediately.
- continue: Skips the current iteration and moves to the next.
- pass: Does nothing, acts as a placeholder.

**Impact of Abnormal Termination**

- Objects created within the loop’s scope are destroyed upon loop exit.
- Abnormal termination introduces exceptions to the normal control flow.

**Best Practices for Loops**
- Every loop should have a normal flow with a single entry and exit point.
- Avoid abnormal terminations when possible to maintain clean control flow.













###**Break Statement**
- The break statement terminates the loop.
- Execution transfers to the statement immediately after the loop.
- When encountered, the loop ends immediately, regardless of the current iteration.
- The rest of the loop body is skipped.
- It doesn’t check the loop's condition; it directly exits the loop.
- Execution resumes with the statement following the loop.
- break only exits the loop in which it’s placed.
- Nested loops: break affects only the current loop, not outer loops.




````python
#Program to show use of break statement to terminate for loop in between.
for num in range(11, 20):
  print(num)
  if num > 15:
    break
#Outside the loop and decision making statement.
print("Finished Task")
````

- The loop’s condition is checked again for the next iteration.
- If the condition is true, the loop starts from the top.

break and continue in Nested Loops

- In nested loops, they will only control the inner loop, not the outer loop.




````python
#Program to add positive integers till user enters 0.
total = 0
complete = False
while not complete:
  amount = eval(input("Input the number (press 0 to quit):"))
  if amount < 0:
    print("Wrong!!!, You entered a negative value",amount)
    continue
  #Skip rest of body for this iteration.
  if amount == 0:
    complete = True
  else:
    total += amount
#Outside the loop and decision making statement.
print("The total number is = ", total)
print("Finished Task")
````

##**User-Defined Functions**
- The function is a standard unit of reuse.
- Functions are essential for large or complex programs.
- They help organize tasks sequentially using a top-down modular approach.
- Simplify debugging due to modularity.
- Reduce code complexity by breaking tasks into smaller, manageable parts.
- The interpreter passes control to the function with arguments.
- The function performs its task and returns control and any result to the interpreter.
- Python allows creating custom functions to extend its functionality.
- These functions act like built-in functions and can be reused or shared.
- A function consists of:
  - Name: Identifies the function.
  - Parameters: Values passed to the function (may be empty).
  - Result: The output produced (could be None).
- Functions can have parameters or no arguments.
- They allow developers to define an interface for users to interact with the code.


Syntax:
```python
def fname(arg1, arg2, ...):
  function body
```
where,
- fname is the actual name of the function stored in Python environment.
- arg1, arg2 are the optional arguments. An argument is like a placeholder. When a function is invoked, we pass a value to the argument. This value is referred to as actual parameter or argument. The argument list refers to the order and number of the arguments of a function. A function can/cannot contain arguments.
- function body contains a collection of statements that define the task of function.
- The function name along with the list of arguments together form the function's signature.

- A function's task is defined upon creation.
- To execute the task, the function must be called.
- Calling a function transfers control to the function.
- The function performs its designated task.
- Control is returned to the main program when the return statement is executed or the last line is reached.
- To call a function, provide the required parameters along with the function name.
- If the function returns a value, store it in a variable.

````python
# Creating functions with arguments and with return statements.

# Function to calculate the area of a square.
def area_square(side):
    return side * side

# Function to calculate the area of a circle.
def area_circle(rad):
    return 3.14 * rad * rad

# Function to calculate the area of a rectangle.
def area_rectangle(length, breadth):
    return length * breadth

# Function to calculate the area of a triangle using Heron's formula.
def area_triangle(a, b, c):
    s = (a + b + c) / 2
    # Correct Heron's formula: √[s(s - a)(s - b)(s - c)]
    area = (s * (s - a) * (s - b) * (s - c)) ** 0.5
    return area

# Calling the function according to user choice.
choice = int(input("Choice: 1-square, 2-circle, 3-rectangle, 4-triangle: "))

if choice == 1:
    side = float(input("Input the length of the square: "))
    area = area_square(side)

elif choice == 2:
    rad = float(input("Input the radius of the circle: "))
    area = area_circle(rad)

elif choice == 3:
    length, breadth = map(float, input("Input the length and breadth of the rectangle (separated by a comma): ").split(','))
    area = area_rectangle(length, breadth)

elif choice == 4:
    a, b, c = map(float, input("Input the three sides of the triangle (separated by a comma): ").split(','))
    area = area_triangle(a, b, c)

else:
    print("Not a valid choice")
    area = None

if area is not None:
    print("The area of the shape is:", area)
````

- The main utility of functions is to create it once and use it for many times. The following example creates a function once and uses it two times to determine whether two numbers ae amicable.
- Two numbers are said to be amicable if the sum of proper divisors of one number plus 1 is equal to the other number. All divisors of a number other than 1 and itself are called proper divisors.
- For example, the numbers 220 and 284 are amicable as the sum of proper divisors of 220 (i.e.) 1, 2, 4, 5, 10, 11, 20, 22, 44, 55 and 110 is equal to 284 and sum of proper divisors of 284 (i.e.) 1, 2, 4, 71 and 142 is 220.

````python
# Function to calculate the sum of factors of a number.
def sum_of_factors(num):
    sum = 1  # Start with 1 because 1 is a factor of every number.
    for i in range(2, num):
        if num % i == 0:
            sum += i
    return sum

# Accepting input from the user.
x, y = eval(input("Enter the two numbers separated by a comma: "))

# Calling the function.
val1 = sum_of_factors(x)
val2 = sum_of_factors(y)

# Check if the numbers are amicable.
if (val1 == y) and (val2 == x):
    print("The numbers are amicable")
else:
    print("The numbers are not amicable")
````

````python
# Function to return the integer value of a Roman numeral character.
def value(roman):
    if roman == 'I':
        return 1
    if roman == 'V':
        return 5
    if roman == 'X':
        return 10
    if roman == 'L':
        return 50
    if roman == 'C':
        return 100
    if roman == 'D':
        return 500
    if roman == 'M':
        return 1000
    return -1  # Return -1 if the character is not a valid Roman numeral.

# Input the Roman numeral from the user.
roman_num = input("Enter the Roman numeral: ").upper()

length = len(roman_num)
ans = 0  # Variable to store the result.

# Loop through the Roman numeral string.
for i in range(length):
    # Get the value of the current Roman numeral character.
    current_value = value(roman_num[i])

    # If there is a next character and its value is greater, subtract the current value.
    if i + 1 < length and value(roman_num[i + 1]) > current_value:
        ans -= current_value
    else:
        ans += current_value

# Display the output.
print("The equivalent integer value of the Roman numeral is:", ans)
````

###**Nesting of Functions**
- The concept of nesting of function is to define one function within other function.
- Due to simple recursive scope rules, inside function is itself invisible outside of its immediately enclosing function, but can see (access) all local objects (data, functions, types, etc.) of its immediately enclosing function as well as of any function(s) which, in turn, encloses that function.

````python
# Program to show nested functions.

# Create two functions.
def greet():
    print("Hello Dear!!!")

def welcome():
    # Calling the nested greet() function inside welcome().
    greet()
    print("Welcome to Python World!!!")

# Calling the single function 'welcome'.
welcome()
````

We can nest functions theoretically to unlimited depth.

````python
# Function to print a greeting message
def greet():
    print("Hello Dear!!!")

# Function to print a welcome message and call greet() function
def welcome():
    greet()
    print("Welcome to Python World!!!")

# Function to check if a number is an Armstrong number
def armstrong(num):
    # Step 1: Find the number of digits in 'num'
    num_digits = len(str(num))  # Convert number to string to easily count digits

    # Step 2: Initialize sum to 0
    sum = 0
    n = num  # Copy the original number for manipulation

    # Step 3: Loop through each digit of the number
    while n > 0:
        ans = n % 10  # Extract the last digit
        sum = sum + (ans ** num_digits)  # Raise the digit to the power of num_digits and add to sum
        n = n // 10  # Remove the last digit

    # Step 4: Check if the sum is equal to the original number
    if sum == num:
        print(f"{num} is an Armstrong number.")
    else:
        print(f"{num} is not an Armstrong number.")


# Main task function that calls armstrong() and prints a closing message
def task():
    armstrong(7136)
    print("Thanks!!!")

# Calling functions
welcome()
task()
print("See You!!!")
````

We can also use return statement inside the function along with the calling of any other function as illustrated in the following example:

````python
# Function to print a welcome message
def welcome():
    print("Welcome to Python World!!!")

# Function to reverse a number, which calls the welcome function
def reverse(num):
    welcome()  # Call the welcome function inside reverse
    reverse_num = 0
    while num > 0:
        remainder = num % 10  # Extract the last digit
        reverse_num = (reverse_num * 10) + remainder  # Build the reversed number
        num = num // 10  # Remove the last digit
    return reverse_num  # Return the reversed number

# Calling the reverse function and storing the result
result = reverse(613)

# Printing the reversed number
print("The reversed number is:", result)
````

It is also possible to pass a string argument in the function and the nesting of functions is also possible for string argument. This is illustrated in the following example:

````python
# Function to determine whether each letter in a string is a vowel or consonant.
def vowel(input_str):
    vowel_set = ['a', 'e', 'i', 'o', 'u']

    for letter in input_str:
        if letter.isalpha():  # Check if the character is a letter
            if letter.lower() in vowel_set:  # Check in lower case
                print("The letter", letter, "is a vowel")
            else:
                print("The letter", letter, "is a consonant")
        else:
            print("The character", letter, "is not an alphabetic letter.")

# Function to determine the length of the string.
def length(input_str):
    count = 0

    for letter in input_str:
        if letter != '':  # Check if the letter is not an empty string
            count += 1  # Increment the count

    return count  # Return the final count

# Main function for accepting user's choice.
def main():
    choice = int(input("Choice 1 for Determining Vowel, 2 for Length: "))
    user_string = input("Input the string: ")

    if choice == 1:
        vowel(user_string)
    elif choice == 2:
        print("The length of the string is:", length(user_string))
    else:
        print("Invalid choice")

# Main Program Loop
valid = "yes"

while valid.lower() == "yes":
    main()
    valid = input("Do you wish to continue? (yes/no): ")

print("Finished Task")
````

##Recursive Functions
A recursive function calls itself from its own body of function. In specific situations, it helps in reducing lines of code to a great extent and is highly useful.

````python
# Creating a recursive function for computing factorial.
def factorial(num):
    if num == 0:  # Check if the number is 0
        return 1  # The factorial of 0 is 1
    else:
        return num * factorial(num - 1)  # Recursive call

# Calling the function
value = int(input("Input the number: "))  # Get user input
print("The factorial of", value, "is:", factorial(value))  # Output the result
````



However, this problem can be solved by changing the scope of the variable to “global” inside a

````python
# Global variable
total = 1000

# Creating the first function
def offer():
    global discount
    discount = 300  # Local variable

    # Inner function to calculate amount
def amount():
    print("The total is:", total)  # Accessing global variable
    print("The amount is:", total - discount)  # Accessing local variable from the outer function

amount()
offer()
````

The following example uses different functions and demonstrates nesting of functions along with
the usage of global variables:

````python
# Global variables
vall = 0
val2 = 0
val3 = 0
ans = 0

# Creating function for displaying options
def show():
    print("Add: Adds three numbers")
    print("Subtract: Subtracts three numbers")
    print("Multiply: Multiplies three numbers")

# Creating function for accepting choice by user
def choice():
    return input("Choose an option (A for Add, S for Subtract, M for Multiply): ")

# Creating function for accepting input
def get_var():
    global vall, val2, val3
    vall = int(input('Value of first number: '))
    val2 = int(input("Value of second number: "))
    val3 = int(input("Value of third number: "))

# Creating addition function
def add():
    global ans
    ans = vall + val2 + val3

# Creating subtract function
def subtract():
    global ans
    ans = vall - val2 - val3

# Creating multiply function
def multiply():
    global ans
    ans = vall * val2 * val3

# Creating an output function
def output():
    print("The total is:", ans)

# Creating main function
def main():
    show()
    userchoice = choice()

    if userchoice == "A" or userchoice == "a":
        get_var()
        add()
        output()
    elif userchoice == "S" or userchoice == "s":
        get_var()
        subtract()
        output()
    elif userchoice == "M" or userchoice == "m":
        get_var()
        multiply()
        output()
    else:
        print("Invalid choice")

# Calling the main function
main()
````
