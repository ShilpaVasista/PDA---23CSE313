### Variables in Python
A Python identifier is a name that is used to identify variables, functions, or any other user-defined item.
## Characteristics of Python Variables

- **Dynamic Typing**: Python is a dynamically typed language, meaning that you do not need to declare the type of a variable when you create it. The type is inferred from the value assigned to it, and it can change during execution. For example:

```python
x = 10       # x is an integer
x = "Hello"  # now x is a string
```

- **Case Sensitivity**: Variable names in Python are case-sensitive. This means that `variable`, `Variable`, and `VARIABLE` would be considered three different variables.

- **Memory Allocation**: When a variable is created, memory is allocated for it based on the data type of the value assigned. The Python interpreter manages memory allocation automatically.

## Naming Rules for Variables

When naming variables in Python, follow these rules:

- Variable names must start with a letter (a-z, A-Z) or an underscore (_).
- They can contain letters, numbers (0-9), and underscores but cannot start with a number.
- Variable names cannot include spaces or special characters (like @, #, $, etc.).
- Reserved keywords (like `if`, `else`, `while`, etc.) cannot be used as variable names.

Examples of valid variable names:

```python
user_age = 20
_userName = "Alice"
score1 = 95
```

Examples of invalid variable names:

```python
1st_score = 50      # cannot start with a number
user-name = "Bob"   # cannot use hyphen
```

## Types of Variables

Python supports various built-in data types that variables can hold:

- **Numeric Types**: Integer values (e.g., `x = 10`)
- **Text Type**: Strings (e.g., `name = "Alice"`)
- **Sequence Types**: Ordered, mutable collections (e.g., `list = [1, 2, 3]`)
- **Mapping Type**: Key-value pairs (e.g., `dictionary = {"key": "value"}`)
- **Set Types**: Unordered collections of unique elements (e.g., `set_var = {1, 2, 3}`)
- **Boolean Type**: Represents truth values (`True`, `False`)

## Variable Scope

Variables in Python have different scopes depending on where they are defined:

- **Global Variables**: Defined outside any function and accessible throughout the program.

```python
global_var = "I am global"

def print_global():
    print(global_var)

print_global()  # Output: I am global
```

- **Local Variables**: Defined within a function and only accessible inside that function.

```python
def print_local():
    local_var = "I am local"
    print(local_var)

print_local()
# print(local_var) would raise an error since local_var is not accessible here.
```

- **Using Parameters**:

```python
print("Python", "is", "fun", sep=", ", end="!")
```

- **Program to print multiple outputs in a single program.**
```
print('Welcome')
print('to')
print('Python World')
```
- **Print with "end" Argument**
  - Sometimes it is convenient to view the output of a single line of printed text over several Python statements.
  - Example: We may compute part of a complicated calculation, print an intermediate result, finish the calculation, and print the final answer with the output all appearing in one line of text.
  - The end argument in print statement allows us to do so.
```python
print('Welcome',end=' ')
print('to',end=' ')
print('Python World',end=' ')
```
  - This will cause the cursor to remain on the same line for the next text. Hence, all the words are printed in the same line.

- **Print with "sep" Argument**
  - The print() function uses a keyword named "sep" to specify the string to use insert between items.
  - The value of "sep" argument allows us to control how the print() function visually separates the argument it displays.
  - The name "sep" stands for separator.
  - By default, the print() function places a single space in between the items it prints.
  - The default value of "sep" is a blank string ' ', which is a string containing single space.
```python
#Program to show the use of "sep" argument in print statement.
var1, var2, var3, var4 = 100, 150, 200, 250
#Printing variables without separator.
print(var1, var2, var3, var4)
#Printing variables using no space as separator.
print(var1, var2, var3, var4, sep='')
#Printing variables using comma as separator.
print(var1, var2, var3, var4, sep = ',')
#Printing variables using colon as separator.
print(var1, var2, var3, var4, sep=':')
#Printing variables using ----- as separator.
print(var1, var2, var3, var4, sep='-----')
```
