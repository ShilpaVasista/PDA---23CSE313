# The matplotlib Library

- Matplotlib is a Python library specializing in the development of two-dimensional charts(including 3D charts);

*  Extreme simplicity in its use

# **Matplotlib Architecture**
- Provides functions and tools for representing and manipulating a Figure (main object) and its internal components.
- Includes event handling and animation capabilities for interactive charts (e.g., triggered by keyboard or mouse events).
- The architecture of matplotlib is logically structured in three layers placed at three different levels
- The communication is unidirectional, that is, each layer is able to communicate with the underlying layer, while the lower cannot do it with the top one.
- The three layers are as follows:
    - scripting
    - artist
    - backend

**Backend Layer**

- In the diagram of the matplotlib architecture the layer that works at the lowest level is the Backend.
- In this layer there are the matplotlib APIs, a set of classes that play the role of implementation of the graphic elements at a low level.
    - FigureCanvas is the object that embodies the concept of drawing area.
    - Renderer is the object that draws on FigureCanvas.
    - Event is the object that handles user inputs (keyboard and mouse events).

**Artist Layer**
- As an intermediate layer we have a layer called Artist.
- All that you can look inside a figure is an Artist object,
that is, all the elements that go to make up a chart such as the title, axis labels, markers, etc., are instances of
the Artist object.
- Each of these instances within the chart plays its role within a hierarchical structure

Screen Shot 2024-12-17 at 09.14.02.png

- There are two Artist classes: **primitive** and **composite**.
- The primitive artists are individual objects that constitute the basic elements to form a graphical
representation in a plot.
- For example a Line2D, or as a geometric figure such as a Rectangle or Circle, or even
pieces of text.
- The composite artists are those graphic elements present in a chart which are composed of several
base elements, namely, the primitive artists.
- Composite artists are for example the Axis, Ticks, Axes, and
figures.

- The three main Artist objects (composite
artists) that are generally used in all implementations:
    - Figure:
      -  Top-level object in the hierarchy.
      - Represents the entire graphical output.
      - Can contain multiple Axes objects.
    - Axes
      - Represents a plot or chart within a Figure.
      - Each Axes belongs to a single Figure.
      - Includes Artist Axis objects (two for 2D, three for 3D).
      - Other components, such as the title, x-label, and y-label, are part of this object.
    - Axis
      - Handles numerical values displayed on Axes.
      - Defines axis limits, tick positions, and tick labels.
      - Tick positions are managed by the Locator object.
      - Tick label formatting is handled by the Formatter object.

**Scripting Layer **
- Best suited for data analysis and visualization tasks.
- Consists of the pyplot module, which provides an easy-to-use scripting interface.

**Matplotlib and NumPy Integration:**

- Despite being a graphical library, Matplotlib relies on NumPy for internal operations.
- Data passed to Matplotlib (such as lists for axes) is automatically converted into NumPy arrays for efficient processing.

**Using NumPy Arrays with Matplotlib:**

- You can directly pass NumPy arrays as input to Matplotlib, which simplifies data handling.
- Data processed by libraries like Pandas can be used directly with Matplotlib without further adjustments.

**Plotting Multiple Trends:**

- You can plot multiple trends (such as a sinusoidal function) on the same plot.
- Use NumPy's arange() to create a series of points on the x-axis.
- Use Python's sin() function from the math module to generate corresponding y-axis values.

```python
import math
import numpy as np
import matplotlib.pyplot as plt

# Create the time array
t = np.arange(0, 2.5, 0.1)

# Use map and convert to list
y1 = list(map(lambda x: math.sin(math.pi * x), t))
y2 = list(map(lambda x: math.sin(math.pi * x + math.pi/2), t))
y3 = list(map(lambda x: math.sin(math.pi * x - math.pi/2), t))

# Plot the data
plt.plot(t, y1, 'b*', label='sin(t)')
plt.plot(t, y2, 'g^', label='sin(t + /2)')
plt.plot(t, y3, 'ys', label='sin(t - /2)')

# Add a title and labels
plt.title('Multiple Sinusoidal Functions')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')

# Show the legend
plt.legend()

# Display the plot
plt.show()
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt

# Generate x-axis data
t = np.arange(0, 2.5, 0.1)  # Array of values from 0 to 2.5 with a step of 0.1

# Generate y-axis data
y1 = list(map(lambda x: math.sin(math.pi * x), t))  # Sinusoidal trend
y2 = list(map(lambda x: math.sin(math.pi * x + math.pi / 2), t))  # Phase-shifted by 90
y3 = list(map(lambda x: math.sin(math.pi * x - math.pi / 2), t))  # Phase-shifted by -90

# Plot all three trends
plt.plot(t, y1, 'b--', label='y1: blue dashed')
plt.plot(t, y2, 'g', label='y2: green solid')
plt.plot(t, y3, 'r-.', label='y3: red dash-dot')

# Add labels, legend, and show plot
plt.xlabel('Time (t)')
plt.ylabel('Values')
plt.title('Sinusoidal Trends with Different Line Styles')
plt.legend()
plt.show()
```
